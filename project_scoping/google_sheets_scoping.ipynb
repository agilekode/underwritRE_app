## create user folder and user model folders 

```python
import gspread
import pandas as pd
from google.oauth2.service_account import Credentials
from googleapiclient.discovery import build

# ‚úÖ Shared config
SERVICE_ACCOUNT_FILE = '../uw_backend/secrets/service_account_key.json'
SCOPES = [
    "https://www.googleapis.com/auth/spreadsheets",
    "https://www.googleapis.com/auth/drive"
]
PARENT_FOLDER_ID = '1xoc6MuOW8ULr3PIucwoEhG0hwpWeIxdk'

# ‚úÖ Load credentials once
creds = Credentials.from_service_account_file(SERVICE_ACCOUNT_FILE, scopes=SCOPES)

# ‚úÖ gspread client for Sheets
gs_client = gspread.authorize(creds)

# ‚úÖ Drive client for creating folders
drive_service = build('drive', 'v3', credentials=creds)

def get_existing_folder_id(service, name, parent_id):
    """Check if a folder with the given name already exists in the parent folder."""
    query = f"mimeType='application/vnd.google-apps.folder' and name='{name}' and '{parent_id}' in parents and trashed = false"
    results = service.files().list(q=query, spaces='drive', fields='files(id, name)').execute()
    folders = results.get('files', [])
    return folders[0]['id'] if folders else None

def create_drive_folder(service, name, parent_id):
    """Create folder only if it doesn't exist already."""
    existing_id = get_existing_folder_id(service, name, parent_id)
    if existing_id:
        print(f"üìÇ Folder already exists: {name} (ID: {existing_id})")
        return existing_id

    file_metadata = {
        'name': name,
        'mimeType': 'application/vnd.google-apps.folder',
        'parents': [parent_id]
    }
    folder = service.files().create(body=file_metadata, fields='id').execute()
    print(f"‚úÖ Created folder: {name} (ID: {folder['id']})")
    return folder['id']

def create_nested_folders(email: str, model_name: str, model_id: str):
    email_folder_id = create_drive_folder(drive_service, email, PARENT_FOLDER_ID)
    full_model_folder_name = f"{model_name} {model_id}"
    model_folder_id = create_drive_folder(drive_service, full_model_folder_name, email_folder_id)
    print(f"üìÅ Final nested folder: {email}/{full_model_folder_name} (ID: {model_folder_id})")
    return model_folder_id

# ‚úÖ Example: Create folders
# create_nested_folders(
#     email='paulgaudin314@gmail.com',
#     model_name='multi family home example 1',
#     model_id='171d3d2b-48ef-4b2e-84b9-735dc65da259'
# )
```

## update the variable in the sheets

```python
from collections import defaultdict

def update_copied_sheet_values(sheet_id, mapped_values, df):
    # Open the copied sheet
    copied_sheet = gs_client.open_by_key(sheet_id)

    # Build mapping: (section, key) -> (sheet_name, cell)
    location_map = {}
    for _, row in df.iterrows():
        loc_str = row['location']
        if isinstance(loc_str, str) and loc_str.startswith('='):
            try:
                sheet_part, cell = loc_str[1:].split('!')
                location_map[(row['section'], row['key'])] = (sheet_part.strip(), cell.strip())
            except ValueError:
                print(f"‚ö†Ô∏è Invalid location format for {row['section']} > {row['key']}: {loc_str}")

    # Prepare a dict: sheet_name -> list of {range, values}
    updates_by_sheet = defaultdict(list)

    for item in mapped_values:
        key = item['field_key']
        value = item['value']
        section = item['section']
        field_type = item.get('field_type')

        sheet_cell = location_map.get((section, key))
        if not sheet_cell:
            print(f"‚ö†Ô∏è No location found for {section} > {key}")
            continue

        # Add % if percent
        if field_type == 'percent' and value:
            try:
                float(value)
                value = f"{value}%"
            except ValueError:
                print(f"‚ö†Ô∏è Could not format percent for {section} > {key}: {value}")

        sheet_name, cell = sheet_cell
        updates_by_sheet[sheet_name].append({
            'range': f"{sheet_name}!{cell}",
            'values': [[value]]
        })

    # Batch update per sheet
    for sheet_name, updates in updates_by_sheet.items():
        try:
            copied_sheet.values_batch_update(
                body={
                    "valueInputOption": "USER_ENTERED",
                    "data": updates
                }
            )
            print(f"‚úÖ Batch updated {len(updates)} cells in '{sheet_name}'")
        except Exception as e:
            print(f"‚ùå Failed batch update for sheet '{sheet_name}': {e}")

update_copied_sheet_values(copied_sheet_id, mapped_values, df)
```

## Get tables for frontend 

```python
import pandas as pd
from IPython.display import display, Markdown

def extract_tables_from_mapping(sheet_id):
    # Open the copied sheet and the Table Mapping worksheet
    copied_sheet = gs_client.open_by_key(sheet_id)
    mapping_ws = copied_sheet.worksheet("Table Mapping")
    mapping_data = mapping_ws.get_all_records()

    for entry in mapping_data:
        table_name = entry.get("table_name")
        location = entry.get("table_location")

        if not location or '!' not in location:
            print(f"‚ö†Ô∏è Skipping invalid location for table '{table_name}': {location}")
            continue

        # Parse sheet name and range
        sheet_name, range_str = location.split('!', 1)

        try:
            # Fetch values from the specified range
            table_values = copied_sheet.worksheet(sheet_name).get(range_str)
            df = pd.DataFrame(table_values)

            # Pretty-print in notebook
            display(Markdown(f"### üìä Table: {table_name} (`{location}`)"))
            display(df)
        except Exception as e:
            print(f"‚ùå Failed to fetch table '{table_name}' from {location}: {e}")

extract_tables_from_mapping(copied_sheet_id)
```

```python
import pandas as pd
from IPython.display import display, Markdown
from googleapiclient.discovery import build
from typing import Optional
!pip3 install Jinja2

def extract_tables_from_mapping_with_styles(sheet_id, creds):
    # Build raw Sheets API client
    sheets_service = build("sheets", "v4", credentials=creds)

    # Get Table Mapping sheet using gspread (still ok for quick read)
    copied_sheet = gs_client.open_by_key(sheet_id)
    mapping_ws = copied_sheet.worksheet("Table Mapping")
    mapping_data = mapping_ws.get_all_records()

    for entry in mapping_data:
        table_name = entry.get("table_name")
        location = entry.get("table_location")

        if not location or '!' not in location:
            print(f"‚ö†Ô∏è Skipping invalid location for table '{table_name}': {location}")
            continue

        sheet_name, range_str = location.split('!', 1)
        range_notation = f"{sheet_name}!{range_str}"

        try:
            # Fetch full cell data with styles
            result = sheets_service.spreadsheets().get(
                spreadsheetId=sheet_id,
                ranges=[range_notation],
                includeGridData=True
            ).execute()

            grid_data = result['sheets'][0]['data'][0]['rowData']
            values, styles = [], []

            for row in grid_data:
                row_values, row_styles = [], []
                for cell in row.get("values", []):
                    val = cell.get("formattedValue", "")
                    fmt = cell.get("effectiveFormat", {})
                    text_fmt = fmt.get("textFormat", {})
                    bg_color = fmt.get("backgroundColor", {})

                    # Extract bold
                    bold = text_fmt.get("bold", False)

                    # Convert background color to CSS RGB
                    def to_rgb(c): return int(c * 255)
                    r = to_rgb(bg_color.get("red", 1))
                    g = to_rgb(bg_color.get("green", 1))
                    b = to_rgb(bg_color.get("blue", 1))

                    style = f"background-color: rgb({r},{g},{b});"
                    if bold:
                        style += " font-weight: bold;"

                    row_values.append(val)
                    row_styles.append(style)

                values.append(row_values)
                styles.append(row_styles)

            # Display with styles
            df = pd.DataFrame(values)
            style_df = pd.DataFrame(styles)
            styled = df.style.apply(lambda _: style_df, axis=None)

            display(Markdown(f"### üìä Table: {table_name} (`{location}`)"))
            display(styled)

        except Exception as e:
            print(f"‚ùå Failed to fetch styled table '{table_name}' from {location}: {e}")
extract_tables_from_mapping_with_styles(copied_sheet_id, creds)
```

## delete mapping sheets then download updated workbook 

```python
import os
import subprocess
from pathlib import Path
from googleapiclient.discovery import build

def export_google_sheet_as_ods(sheet_id: str, output_path: str = "exported_model.ods"):
    drive_export_service = build('drive', 'v3', credentials=creds)

    request = drive_export_service.files().export_media(
        fileId=sheet_id,
        mimeType='application/x-vnd.oasis.opendocument.spreadsheet'  # ODS format
    )

    with open(output_path, 'wb') as f:
        f.write(request.execute())

    print(f"‚úÖ ODS file saved to: {os.path.abspath(output_path)}")
    return output_path

from typing import Optional


def convert_with_soffice(input_path: str, output_path: str) -> Optional[str]:
    input_path = Path(input_path)
    output_path = Path(output_path)
    output_dir = output_path.parent
    expected_output = output_dir / (input_path.stem + ".xlsx")

    try:
        subprocess.run(
            [
                "soffice", "--headless", "--convert-to", "xlsx",
                "--outdir", str(output_dir), str(input_path)
            ],
            check=True
        )

        if expected_output.exists():
            # Rename to the desired output_path
            expected_output.rename(output_path)
            print(f"‚úÖ Converted with soffice and renamed: {output_path.resolve()}")
            return str(output_path)
        else:
            print("‚ö†Ô∏è soffice did not create the expected .xlsx file")
            return None
    except subprocess.CalledProcessError as e:
        print(f"‚ùå soffice failed: {e}")
        return None

# ‚úÖ Run the full export + dual conversion
ods_filename = f"example output version {VERSION}.ods"
ods_path = export_google_sheet_as_ods(copied_sheet_id, ods_filename)

xlsx_soffice_path = ods_path.replace(".ods", "_soffice.xlsx")

convert_with_soffice(ods_path, xlsx_soffice_path)
```
